# Text pre-processing

```python
import nltk
from nltk import word_tokenize 
```

```python
text1 = textDf["Publication"].tolist()
year1 = textDf["Year"].tolist()
```

```python
# the word_tokenize function
text_tokenised = []
year_final = []

for j in range(len(text1)):
    try:
        # Tokenise text
        tokens = word_tokenize(text1[j].lower()) 
        if(len(tokens)>1): 
            text_tokenised.append(tokens) 
            year_final.append(year1[j]) 
    except:
        pass
print("[DONE]\n")
print("Total text:",len(text_tokenised))
print("Total labels:",len(year_final),"\n")
print(text_tokenised[0])
```

```python
# Import the stop words lists from NLTK
from nltk.corpus import stopwords 
# Import the re package
import re
```

```python
# Load the stop words list for English in variable
stopwords_english = stopwords.words('english') 
# Create copy of text_tokenised
text_preprocessed = text_tokenised 
print("Preprocessing text..",end="")
for a in range(len(text_tokenised)):
    new_tokens = []
    for word in text_tokenised[a]:
        # Remove hyphens from words, e.g. lower-case->lowercase
        word = word.replace("-","") 
        # Remove dots from words to normalise abbreviations
        word = word.replace(".","") 
        # Select only tokens that consist of letters from a to z
        regex_check = re.match("[a-z]+", word)
        if(regex_check!=None) and word not in stopwords_english:
            if(regex_check.group()==word):
                new_tokens.append(word)
    text_preprocessed[a] = new_tokens
print("[DONE]\n")
# Check if pre-processing led to any empty text
for b in range(len(text_preprocessed)):
    if(len(text_preprocessed[b])==0):
        print("text",b,"is empty!")
print(text_preprocessed[0]) 
```

```python
#Join text words list
dataset = []
for c in range(len(text_preprocessed)):
    # Join words with an empty space between them
    texT = " ".join(text_preprocessed[c]) 
    dataset.append(texT)
print(dataset[0])
textt = []
labelss = []
for k in range(len(dataset)):
    if len(dataset[k]) > 0:
        textt.append(dataset[k])
        labelss.append(year_final[k])
textt1 = textt
labelss1 = labelss
data1 = {"Publication" : textt1 , "Year" :labelss1}
df3 = pd.DataFrame(data1)
print(df3.head())
df3.to_csv('peacedataset.csv',index=False) 
```

```python
from sklearn.feature_extraction.text import TfidfTransformer  
from sklearn.feature_extraction.text import CountVectorizer 
```

```python
def tokenize(text):
    try:
        regex = re.compile('[' +re.escape(string.punctuation) + '0-9\\r\\t\\n]')
        text =regex.sub(" ",text)
                           
        tokens_ = [word_tokenize(s) for s in sent_tokenize(text)]
        tokens = []
        for token_by_sent in tokens_:
            tokens+= token_by_sent
        tokens = list(filter(lambda t:t.lower() not in stopwords_english,tokens))
        filtered_tokens = [w for w in tokens if re.search('[a-zA-Z]',w)]
        filtered_tokens = [w.lower() for w in filtered_tokens if len(w)>=3]
        
        return filtered_tokens
    
    except TypeError as e:print(text,e)
```
